## Why

Conditional expectation/probability is how measure-theoretic probability "zooms in" on partial information. If a full experiment is given by $\OmegaFP$, then learning only $\scriptG\isasubsetof\scriptF$ (a coarser view of outcomes), should update both expectation and events' probabilities.

Conditioning formalizes this update so that:

- integrals "localize" to what $\scriptG$ can see,
- randomness splits cleanly into "explained by $\scriptG$" and leftover noise.
- Joint laws can be factored into marginals $\times$ conditionals (i.e. disintegration)â€”the backbone of Bayesian inference comma, Markov processes, and regression.


## How

### Measure-theoretic definition

Let $\OmegaFP$ be a probability space, $X\bein\L1wrtP$ and $\G\beasubsigmaalgebraof\F$.

**Conditional Expectation**. A random variable $Z$ is a version of $\E[X|G]$ if

- $Z$ is $\G$-measureable, and
- $\int_G{Z\,d\P}=\int_G{X\,d\P}$ for all $G\in\scriptG$
